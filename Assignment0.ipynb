```python
##Problem-1
count = 0
import string
df = dict.fromkeys(string.ascii_lowercase, 0)
path = "/content/drive/My Drive/ML/test.txt"
with open(path,'r')as f:
  for line in f:
    newstring = ''
    newstring = line  
    newstring = newstring.replace(',', '')
    newstring = newstring.replace(' ','')
    newstring = newstring.replace('\n','')
    for letter in newstring:
        if (letter.isupper())==True:
          new_letter = letter.lower()
        else: 
          new_letter = letter
        df[new_letter] += 1 
list1 = list(df.values())
df_ar = np.array(list1)
df_summation = np.sum(df_ar)
#print(type(df))
#print(type(df_ar))
print(df)
```

```python
plt.bar(df.keys(), df_ar/df_summation, color='b')
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.title('Normalized Histogram')
plt.show()
```

```python
##Problem-2
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/drive/My Drive/ML/test.png")
img1 = np.copy(img)
cv2_imshow(img)
```

```python
p1 = np.array([255,0,0])
p2 = np.array([0,255,0])
p3 = np.array([0,0,255])
distance1 = np.linalg.norm(img-p1,axis=2)
distance2 = np.linalg.norm(img-p2,axis=2)
distance3 = np.linalg.norm(img-p3,axis=2)
#cv2_imshow(distance1)
#cv2_imshow(distance2)
#cv2_imshow(distance3)

for i in range(img.shape[0]):
  for j in range (img.shape[1]):
    if (distance1[i,j]<distance2[i,j]) and (distance1[i,j]<distance3[i,j]):
      img1[i,j,:] = p1
    elif (distance2[i,j]<distance1[i,j]) and (distance2[i,j]<distance3[i,j]):
      img1[i,j,:] = p2
    else:
      img1[i,j,:] = p3
cv2_imshow(img1)
cv2.imwrite('output1.png',img1)

```

```python
img2 = np.copy(img)
ul = img.shape[0]
ur = img.shape[1]
start = (img.shape[0] - 50)//2
endi = start +50
startj = (img.shape[1]-50)//2
endj = startj+50
#print(type(ul))
for i in range(ul):
  for j in range(ur):
    if (i>= start) and (i<=endi)and (j>=startj) and (j<=endj):
      img2[i,j,:] =[0,0,0]
cv2_imshow(img2)
cv2.imwrite('output2.png',img2)
```
